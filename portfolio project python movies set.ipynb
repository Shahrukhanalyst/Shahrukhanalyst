```python
#import all requries libries
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import seaborn as sns
%matplotlib inline
df=pd.read_csv('movies.csv')
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (15,8)
```

```python
#read csv file

df=pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
#lets see the missing data
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format (col,pct_missing))
```

```python
#data types of columns
df.dtypes
```

```python
#changes the data type
df['gross']=df['gross'].astype('int64')
df['budget']=df['gross'].astype('int64')
df['votes']=df['votes'].astype('int64')
```

```python
#valueError :cannot convert non-finite values to integer


df=df.fillna(0)
```

```python
df.dtypes
```

```python
#Check for missing values
df.isna().sum()
```

```python
#Drop rows with missing data
cols = ['rating', 'released', 'score', 'votes', 'writer', 'star', 'country', 'budget', 'gross', 'company', 'runtime']

for col in cols:
    df.dropna(subset = [col], inplace=True)
df.reset_index()
df.head()
```

```python
##pd.set_option('display.max_rows',None)
```

```python
#sort data by gross view top 10 by gross
top_10_M=df.sort_values(by=['gross'],inplace=False,ascending=False)
top_10_M=top_10_M.head(10)
top_10_M
```

```python
##country indian
india=df.loc[(df['country']=='India')]
```

```python
#sort indian movies data by gross view top 10
top_10_India=india.sort_values(by=['gross'],inplace=False,ascending=False)
top_10_India=top_10_India.head(10)
top_10_India
```

```python
#top genre accourdimg to gross

genre_gross = df.groupby(['genre'], as_index=False)['gross'].sum().sort_values(by='gross', ascending=False)
genre_gross.reset_index()
genre_gross
```

```python
#top 10 weiters by movies

writers = df.groupby(['writer'], as_index=False)['name'].count().sort_values(by='name', ascending=False)
writers.rename(columns={'name':'movies'}, inplace=True)
top_writers = writers.head(10)
top_writers
```

```python
#Correlations of numeric features from our dataset
cor = df.corr()
sns.heatmap(cor, annot=True, vmin=-1)
plt.title('Movie Correlations for Numeric Features', size=15)
plt.xlabel('Features', size=12)
plt.ylabel('Features', size=12)
plt.show()
```

```python
#Scatter Plot illustarting budget vs gross
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.title('Budget vs Gross Earnings', size=20)
plt.xlabel('Film Budget (millions)', size=15)
plt.ylabel('Gross Earnings (millions)', size=15)

plt.show()
```

```python
#Convert our categorical features into numerical values
df_n = df.copy()
for col in df_n.columns:
    if df_n[col].dtype == 'object':
        df_n[col] = df_n[col].astype('category').cat.codes
df_n.head()
```

```python
#top score movies

df_scoring = df.sort_values(by='score', ascending=False)
top_rated = df_scoring[df_scoring['score']>8.6]
top_rated
```

```python
##Movie Correlations

full_correlactions = df_n.corr()
sns.heatmap(full_correlactions, annot=True, vmin=-1)
plt.title('Movie Correlations for Numeric Features', size=15)
plt.xlabel('Features', size=12)
plt.ylabel('Features', size=12)
plt.show()
```

```python
# bar plot top 10 movies by gross

top_10_M.plot(kind='bar', x='name', y=['budget','gross'], color={'budget':'black', 'gross':'gray'}, figsize=(12,10))
plt.title('Budget and Gross Highest Grossing', size=20)
plt.xlabel('Amount(millions)', size=15)
plt.ylabel('Movie Title', size=15)
plt.show()
```

```python
top_10_India.plot(kind='barh', x='name', y=['budget','gross'], color={'budget':'purple', 'gross':'green'}, figsize=(12,10))
plt.title('Budget and Gross Highest Grossing', size=20)
plt.xlabel('Amount(millions)', size=15)
plt.ylabel('Movie Title', size=15)
plt.show()
```

```python
genre_gross.plot(kind='bar', x='genre', y='gross', color='dodgerblue', figsize=(12,8))
plt.title('Budget and Gross Earnings for Highest Grossing Companies', size=25, y=1.05)
plt.xlabel('Company', size=15)
plt.ylabel('Amount(100 millions)', size=15)
plt.show()
```

```python
top_writers.plot(kind='bar', x='writer', y='movies', color='dodgerblue', figsize=(12,8))
plt.title('Writer vs Number of Movies', size=25, y=1.05)
plt.xlabel('Writer', size=15)
plt.xticks(rotation=35)
plt.ylabel('Number of Movies', size=15)
plt.show()
```
